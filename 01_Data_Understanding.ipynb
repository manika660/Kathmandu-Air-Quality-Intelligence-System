##Interpretation:

The PM2.5 concentration varies significantly over time. Higher pollution levels are observed during the winter months, while lower concentrations occur during the monsoon season. Several sudden spikes indicate short-term pollution events, suggesting that both seasonal and environmental factors influence air quality in Kathmandu.

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Matrix")
plt.show()
```

##The correlation heatmap shows that PM10 has the strongest positive correlation with PM2.5 (0.95), indicating that these two particulate pollutants increase together. Sulphur dioxide (0.75) and carbon monoxide (0.62) also exhibit strong positive relationships with PM2.5, while temperature (-0.52) shows a moderate negative correlation, suggesting that PM2.5 concentrations tend to decrease as temperature rises. Variables such as surface pressure and ozone have weak linear correlations with PM2.5. These findings help identify the most influential features for the air quality prediction model.

```python
X = df.drop(["pm2_5", "year"], axis=1)

y = df["pm2_5"]
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import joblib

joblib.dump(scaler, "scaler.pkl")
```

```python
processed = pd.DataFrame(X_scaled, columns=X.columns)

processed["pm2_5"] = y.values

processed.to_csv("processed_air_quality.csv", index=False)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Create model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.5)

plt.xlabel("Actual PM2.5")
plt.ylabel("Predicted PM2.5")
plt.title("Actual vs Predicted PM2.5")

plt.show()
```

```python
mae = mean_absolute_error(y_test, y_pred_dt)
mse = mean_squared_error(y_test, y_pred_dt)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_dt)
print("Decision Tree Performance")
print("------------------")
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:",rmse)
print("R^2 Score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
mae_rf = mean_absolute_error(y_test, y_pred_rf)

mse_rf = mean_squared_error(y_test, y_pred_rf)

rmse_rf = np.sqrt(mse_rf)

r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Performance")
print("---------------------------")

print("MAE :", mae_rf)
print("MSE :", mse_rf)
print("RMSE:", rmse_rf)
print("R² Score:", r2_rf)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_rf, alpha=0.5)
plt.xlabel("Actual PM2.5")
plt.ylabel("predicted PM2.5")
plt.title("Random Forest : Actual vs predicted PM2.5")
plt.grid(True)
plt.show()
```

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
})

importance = importance.sort_values(
    by = "Importance",
    ascending = False
)
print(importance)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.barh(importance["Feature"], importance["Importance"])
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("Feature Importance")
plt.show()
```

```python
sample = [[
    120,     # pm10
    1600,    # carbon_monoxide
    30,      # nitrogen_dioxide
    12,      # sulphur_dioxide
    60,      # ozone
    18,      # temperature
    70,      # humidity
    2.5,     # wind_speed
    180,     # wind_direction
    875,     # surface_pressure
    1,       # month
    5,       # day
    12,      # hour
    2        # day_of_week
]]

sample = scaler.transform(sample)

prediction = rf_model.predict(sample)

print("Predicted PM2.5:", prediction[0])
```

```python
def get_aqi_category(pm25):
    if pm25 <= 12:
        return " Good"
    elif pm25 <= 35.4:
        return " Moderate"
    elif pm25 <= 55.4:
        return "Unhealthy for Sensitive Groups"
    elif pm25 <= 150.4:
        return " Unhealthy"
    elif pm25 <= 250.4:
        return " Very Unhealthy"
    else:
        return " Hazardous"
```

```python
category = get_aqi_category(prediction[0])

print(f"Predicted PM2.5 : {prediction[0]:.2f} µg/m³")
print(f"Air Quality     : {category}")
```
